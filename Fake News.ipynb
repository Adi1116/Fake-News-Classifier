```python
#IMPOTING ALL NECESSARY LIBRARIES FOR MODEL EVALUTAION AND TRAINING
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("E:\ANACONDA\Fake News.csv")
df
```

```python
#HANDLING AND ANALYSING THE DATAFRAME
df.describe()
```

```python
#REMOVING THE NAN VALUES FROM DATASET
df = df.dropna()
df
```

```python
#REMOVING THE CLASSIFIER COLUMN 'LABEL' FROM INDEPENDENT DATA
x = df.drop('label', axis = 1)
y = df['label']
```

```python
print(x.shape)
print(y.shape)
```

```python
#MAKING A COPY OF INDEPENDENT DATA FOR PREPROCESSING
fake = x.copy()

#RESETTING THE INDEXES TO ORIGINAL VALUE
fake.reset_index(inplace=True)
```

```python
#IMPORTING NLTK MODULES FOR TEXT PREPROCESSING
import nltk
import re
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
#APPLYING THE REQUIRED TEXT PREPROCESSING AND STORING THE NEW TEXT INTO CORPUS
ps = PorterStemmer()
corpus = []
for i in range(0, len(fake)):
    review = re.sub('[^a-zA-Z]', ' ', str(fake['title'][i]))
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in set (stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
#IMPOTING ALL NECESSARY LIBRARIES AND MODULES FOR MODEL EVALUTAION AND TRAINING
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
```

```python
#CONVERTING TEXT INTO ONE HOT REPRESENTATION 
voc_size = 1000
onehot_rep = [one_hot(words, voc_size) for words in corpus]
onehot_rep
```

```python
#DOING WORD EMBEDDING USING PAD SEQUENCE
sent_length = 20
emb_doc = pad_sequences(onehot_rep, padding = 'pre', maxlen = sent_length)
emb_doc[0]
```

```python
#APPLYING THE NEURAL NETWORK MODEL USING EMBEDDING, LSTM, DROPOUT, AND DENSE
emb_vec_feat = 20
model = Sequential()
model.add(Embedding(voc_size, emb_vec_feat, input_length = sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
print(model.summary())
```

```python
#CONVERSION OF EMBEDDED DATA INTO NUMPY ARRAY
x_final=np.array(emb_doc)
y_final=np.array(y)
print(x_final.shape)
print(y_final.shape)
```

```python
#SPLITTING THE AVAILABLE DATA INTO TRAIN AND TEST
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_final, y_final, test_size = 0.33, random_state = 1)
```

```python
#APPLYING THE MODEL ON TRAINING SET
model.fit(x_train, y_train, validation_data = (x_test, y_test), epochs = 10, batch_size = 100)
```

```python
#MAKING OUT PREDICTIONS FOR INDEPENDENT TEST DATA
y_pred = model.predict(x_test)
```

```python
y_pred = np.rint(y_pred)
```

```python
#CALCULATING THE CONFUSION MATRIX AND GETTING THE ACCURACY
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
